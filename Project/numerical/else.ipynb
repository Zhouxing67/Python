```python
import sympy
x, y, z = sympy.symbols('x y z')
#原函数：x**3 - x**2 - x - 1
f_x = x**3 - x**2 - x - 1
#导函数：3*x**2 - 2*x - 1
f_x_d = sympy.diff(f_x,x)
#构造迭代式：x - (x**3 - x**2 - x - 1)/(3*x**2 - 2*x - 1)
x_itr = x - f_x / f_x_d
#规定迭代次数为10次,X0=1.8
x_cur=1.8
for i in range(10):
    x_next=x_itr.evalf(subs={x:x_cur})
    print(f"第{i+1}次迭代, 当前值x_cur={x_cur}, 迭代后的值x_next={x_next}")
    x_cur=x_next

        
import sympy
x, y, z = sympy.symbols('x y z')
#原函数：sin(x) - cos(x) + 0.5
f_x = sympy.sin(x) - sympy.cos(x) + 0.5
#导函数：sin(x) + cos(x)
f_x_d = sympy.diff(f_x,x)
#构造迭代式：x - (sin(x) - cos(x) + 0.5)/(sin(x) + cos(x))
x_itr = x - f_x / f_x_d
#规定迭代次数为10次,X0=pi/4
x_cur=sympy.pi/4
for i in range(10):
    x_next=x_itr.evalf(subs={x:x_cur})
    print(f"第{i+1}次迭代, 当前值x_cur={x_cur}, 迭代后的值x_next={x_next}")
    x_cur=x_next
```

```python
import numpy as np
A = np.array([[4.0, 3.0, 0.0],
                [3.0, 4.0, -1.0],
                [0.0, -1.0, 4.0]])
U = np.tril(A, 0) - A
L = np.triu(A, 0) - A
D = A + (L + U )
D_ = np.linalg.inv(D)
b = np.array([[24], [30], [-24]])

#迭代法的一般形式 ： X = B * X + f
def get_X(B, f, X0):
    print(f"初始值为{X0.T}")
    X_cur = X0
    for i in range(1000):
        X_next = B @ X_cur + f
        X_cur = np.copy(X_next)
        if i < 10 :
            print(f"第{i + 1}次迭代后的值为{X_next.T}")
    return X_next
#雅各比迭代法：
x0 =  np.array([[0], [0], [0]])
x1 =  np.array([[1], [1], [1]])
B1 = D_ @ (L + U);  f1 = D_ @ b
X1 = get_X(B1, f1, x0)
print(f"雅各比迭代法,初值为[[0, 0, 0]]时 迭代一千次的最终结果 :                     {X1.T}")
X1 = get_X(B1, f1, x1)
print(f"雅各比迭代法,初值为[[1, 1, 1]]时 迭代一千次的最终结果 :                     {X1.T}")

#高斯-塞德尔迭代法
B2 = np.linalg.inv(D - L) @ U;  f2 = np.linalg.inv(D - L) @ b
X2 = get_X(B2, f2, x0)
print(f"高斯-塞德尔迭代法,初值为[[0, 0, 0]]时 迭代一千次的最终结果 :                     {X2.T}")
X2 = get_X(B2, f2, x1)
print(f"高斯-塞德尔迭代法,初值为[[1, 1, 1]]时 迭代一千次的最终结果 :                     {X2.T}")


#(逐次)超松弛(SOR)迭代法 w = 1.22
w = 1.22
B3 = np.linalg.inv(D - w * L) @ (w * U + (1 - w) * D);  f3 = w * np.linalg.inv(D - w * L) @ b
X3 = get_X(B3, f3, x0)
print(f"(逐次)超松弛(SOR)迭代法 w = 1.22,初值为[[0, 0, 0]]时 迭代一千次的最终结果 :                     {X3.T}")
X3 = get_X(B3, f3, x0)
print(f"(逐次)超松弛(SOR)迭代法 w = 1.22,初值为[[1, 1, 1]]时 迭代一千次的最终结果 :                     {X3.T}")


#(逐次)超松弛(SOR)迭代法 w = 1.8
w = 1.8
B4 = np.linalg.inv(D - w * L) @ (w * U + (1 - w) * D);  f4 = w * np.linalg.inv(D - w * L) @ b
X4 = get_X(B4, f4, x0)
print(f"(逐次)超松弛(SOR)迭代法 w = 1.8,初值为[[0, 0, 0]]时 迭代一千次的最终结果 :                     {X4.T}")
X4 = get_X(B4, f4, x1)
print(f"(逐次)超松弛(SOR)迭代法 w = 1.8,初值为[[1, 1, 1]]时 迭代一千次的最终结果 :                     {X4.T}")
```
